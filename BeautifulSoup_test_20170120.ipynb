```python
import bs4

html = '''<html><body>
<h1>Title</h1>
<p>Hello, I'm John.</p>
</body></html>'''

soup = bs4.BeautifulSoup(html, 'html.parser')
print(type(soup.p))  #=> <class 'bs4.element.Tag'>
print(soup.p.name)   #=> 'p'
print(soup.p.text)   #=> 'Hello, I'm John.'
```

```python
import requests
import bs4
 
def main():
    r = requests.get("http://developerexcuses.com/")
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    
    print(soup.select("html body div.wrapper center a")[0].text)
    return soup


bs = main()
print(bs)
```
